```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('stock_data.csv')
open_prices = data['Open'].values
high_prices = data['High'].values
low_prices = data['Low'].values
volume_prices = data['Volume'].values
```

```python
#visualize the data
plt.scatter(open_prices, high_prices, marker ='o', color='magenta')
plt.title('Is Open good predictor of days High prices')
plt.xlabel('opening stock')
plt.ylabel('high price')
plt.show()
```

```python
#visualize the data
plt.scatter(open_prices, low_prices, marker ='o', color='pink')
plt.title('Is Open good predictor of days Low prices')
plt.xlabel('opening stock')
plt.ylabel('low price')
plt.show()
```

```python
#visualize the data
plt.scatter(volume_prices, high_prices, marker ='o', color='blue')
plt.title('higher trading activity is associated with higher daily highs.')
plt.xlabel('volume')
plt.ylabel('high price')
plt.show()
```

```python
#data visualization woaw
plt.scatter(volume_prices, low_prices, marker ='o', color='blue')
plt.title('whether higher trading activity is linked to deeper price drops')
plt.xlabel('volume')
plt.ylabel('high price')
plt.show()
```

```python
# Can Open explain High?
x1_train, x1_test, y1_train, y1_test=train_test_split(open_prices, high_prices, test_size=0.2, random_state=42)
```

```python
#Reshape the data 
x1_train = x1_train.reshape(-1, 1)
x1_test = x1_test.reshape(-1, 1)

model = LinearRegression()
model.fit(x1_train, y1_train)
```

```python
# Predict prices for the test set
predictions = model.predict(x1_test)

# Visualize the predictions
plt.scatter(x1_test, y1_test, marker='o', color='blue', label='Actual Prices')
plt.plot(x1_test, predictions, color='red', linewidth=2, label='Predicted Prices')
plt.title('Open good predictor of days High prices with Linear Regression')
plt.xlabel('open prices')
plt.ylabel('high prices')
plt.legend()
plt.show()
```

```python
# Can Open explain Low?
x2_train, x2_test, y2_train, y2_test=train_test_split(open_prices, low_prices, test_size=0.2, random_state=42)
```

```python
# Reshape the data 
x2_train = x2_train.reshape(-1, 1)
x2_test = x2_test.reshape(-1, 1)

model = LinearRegression()
model.fit(x2_train, y2_train)
```

```python
predictions = model.predict(x2_test)

plt.scatter(x2_test, y2_test, marker='o', color='green', label='Actual Prices')
plt.plot(x2_test, predictions, color='red', linewidth=2, label='Predicted Prices')
plt.title('Open good predictor of days low prices with Linear Regression')
plt.xlabel('open prices')
plt.ylabel('low prices')
plt.legend()
plt.show()
```
